```python
import pandas as pd
import numpy as np
```

```python
df_literacy=pd.read_excel(r'C:\Users\Deepshika\Documents\Literacy_Rate_finalf.xlsx')
df_literacy
```

```python
df_literacy['State/UT'] = df_literacy['State/UT'].str.upper()
df_literacy = df_literacy.rename(columns={'State/UT': 'STATE/UT'})
df_literacy = df_literacy.rename(columns={'Year': 'YEAR'})
```

```python
df_literacy
```

```python
df_pop=pd.read_excel(r'C:\Users\Deepshika\Desktop\poplution_yearwise_finalff.xlsx')
df_pop['STATE/UT'] = df_pop['STATE/UT'].str.upper()

df_pop
```

```python
df_pop=df_pop.groupby(['STATE/UT','YEAR']).mean()
df_pop=df_pop.reset_index()
df_pop
```

```python
df_area=pd.read_excel(r'C:\Users\Deepshika\Desktop\area_of_states_final.xlsx')
df_area.sort_values(by='Area(sq km)')
```

```python
df_area['STATE/UT'] = df_area['STATE/UT'].str.upper()
```

```python

df_pop_area = df_pop.merge(df_area, on=['STATE/UT'], how='outer')
df_pop_area
```

```python
#df_pop_area.isnull().sum()
```

```python
df_pop_area_literacy=df_pop_area.merge(df_literacy, on=['STATE/UT','YEAR'], how='outer')
```

```python
df_pop_area_literacy=df_pop_area_literacy.dropna()
df_pop_area_literacy
```

```python
df_pop_area_literacy.to_csv('PHASE1_OUTPUT.csv', index=False)
```
